```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# Load dataset (Mall Customers Data)
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:, [3, 4]].values  # Selecting Annual Income & Spending Score

# Standardizing the data (Important for DBSCAN)
X = StandardScaler().fit_transform(X)

# Applying DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Adjust eps & min_samples as needed
labels = dbscan.fit_predict(X)

# Visualizing the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=50)
plt.title("DBSCAN Clustering of Customers")
plt.xlabel("Annual Income (Standardized)")
plt.ylabel("Spending Score (Standardized)")
plt.show()
```

```python
# Finding outliers (Label -1 means outlier)
outliers = X[labels == -1]
# Printing the number of outliers
print(f"Total Outliers Detected: {len(outliers)}")

# Visualizing the clusters with outliers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=50, label="Clusters")
plt.scatter(outliers[:, 0], outliers[:, 1], c='black', s=100, marker='x', label="Outliers")  # Marking outliers
plt.title("DBSCAN Clustering with Outliers")
plt.xlabel("Annual Income (Standardized)")
plt.ylabel("Spending Score (Standardized)")
plt.legend()
plt.show()
```
